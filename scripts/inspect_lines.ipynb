```python
import numpy as np
import astropy.io.fits as fits
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import os
homedir = os.path.expanduser('~')
```

```python
# Load molecular line list
molist = ['H2O', 'CO', 'CH4']
z =      [    3,    2,     1]
mols = {}
```

```python
for name in molist:
    mol = np.loadtxt(f"{homedir}/uoedrive/data/exomol/{name}_abs.txt")
    wavenumebers = np.arange(714, 11999.1, 0.1)
    x = 1e4 / wavenumebers
    y = mol  # normalize intensity
    mols[name] = np.array([x, y])
```

```python
plt.figure(figsize=(20, 4))
plt.title("Exomol cross sections")
for i, mol in enumerate(['H2O', 'CO', 'CH4']):
    print(mol, z[i])
    plt.plot(mols[mol][0], mols[mol][1], label=mol, alpha=0.5, zorder=5-z[i])
    plt.axvspan(1.90, 2.52, color='gray', alpha=0.1)
    plt.axvspan(2.77, 4.19, color='gray', alpha=0.1)

plt.xlim(3, 5)
#plt.ylim(-0.1e-18, 2.5e-18)
plt.yscale('log')
plt.ylim(1e-30, 5e-18)
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(0.05))
plt.gca().xaxis.set_minor_locator(ticker.MultipleLocator(0.01))
plt.legend();
```
